## Analyzing Profitability of Parent Companies

```python
#Creating a profit_pct column for calculating percentage profit on each film
df_raw['profit_pct'] = (df_raw['gross'] - df_raw['budget'])/df_raw['gross']
```

```python
#Which were the most profitable companies every year?
df_raw_mean = df_raw.groupby(['year','parent_company'], as_index = False).mean().sort_values(by=['year','profit_pct'], ascending=False)
df_raw_mean.count()
```

```python
df_raw_mean = df_raw_mean.reset_index(drop=True)
df_raw_mean[df_raw_mean['year'] >= 2015]
```

```python
#Creating a yearwise Index for df_raw_mean - column "rank"
df_raw_mean['rank'] = np.random.randint(1, size = len(df_raw_mean['year']))
len(df_raw_mean['year'])
```

```python
#populating the "rank" column with index resetting every year
curr_year = 2020
ind = 0
ctr = 0

while(ind <= 2927):
    if df_raw_mean['year'][ind] == curr_year:
        df_raw_mean.loc[ind, 'rank'] = ctr
        ctr = ctr + 1
        ind = ind + 1
    else:
        curr_year = curr_year - 1
        ctr = 0
```

```python
df_raw_mean
```

```python
#Looking at only top 3 highest profitable companies each year from 2015 to 2020 (The rank column makes it pretty convenient to do this)
df_raw_mean[(df_raw_mean['year'] >= 2015) & (df_raw_mean['rank'] < 3)]
```

```python
#high correlation between budget and gross.. implies majority of companies are on average profitable in their production
df_raw_mean[df_raw_mean['year'] >= 2010].corr(method='pearson')
```

```python
#visualizing the correlation between avg budget and avg gross of companies between 2010 and 2020
import seaborn as sns
sns.set(rc = {'figure.figsize':(10,8)})
with sns.axes_style("whitegrid"):
    sns.regplot(x=df_raw_mean[df_raw_mean['year'] >= 2010]['budget'], y=df_raw_mean[df_raw_mean['year'] >= 2010]['gross'], line_kws={'color':'red'})
```

```python
#Question: Are companies with higher avg. IMDB ratings higher grossing?

#Most of the higher avg gross companies have a higher average IMDB rating as well(between 6 - 9)
#I am filtering out movies with lesser than 500 votes to avoid outliers 
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot()

x = df_raw_mean[(df_raw_mean['votes'] >= 500)]['score']
y = df_raw_mean[(df_raw_mean['votes'] >= 500)]['gross']

plt.xlabel('Average IMDB Score')
plt.ylabel('Average Gross')
plt.title('IMDB Score Vs Gross')

# ax.xaxis.label.set_color('white')
# ax.tick_params(axis='x', colors='white')
# ax.yaxis.label.set_color('white')
# ax.tick_params(axis='y',colors="white") 
# ax.title.set_color('white')

plt.scatter(x,y)
plt.show()
```

```python
#Visualizing correlation between gross and year in the movie industry from 1980 to 2020
#the Film Production industry seems to have undergone tremendous growth in this period

reg = sns.regplot(x=yearly_avg['year'], y=yearly_avg['gross'], line_kws={'color':'red'}, scatter_kws={'color':'green'})

reg.set_xlabel('Year')
reg.set_ylabel('Gross (USD)')

reg.set_xticks(years_x_axis)
reg.set_xticklabels(years_x_axis, rotation=90)

plt.show()
```

### [Extra] Analyzing whether "Dump Months" really do exist, based on Gross of Movies during these months

"A dump month is the period of the year where the commercial and critical expectations for films are exceedingly low. Audiences tend to be at their busiest during these periods, keeping them away from the theaters. As a result, studios will spend this time “dumping” the movies that have been deemed lesser. This could be because they were received poorly by test audiences, or because the studio felt that they have a whole host of movies which should be released during more financially lucrative months, such as June and December. This results in there being an excess of very mediocre films during the January period. 

Other <i>dump months besides <b>January</b> include <b>February</b>, and sometimes <b>August</b> and <b>September</b></i>"<br><br>
[ source: https://movies.mxdwn.com/feature/dump-months-what-are-they-and-what-do-they-entail/ ]

```python
#Lets check if the above is true, based on this dataset
#Creating a month column
df_raw['month'] = 'month'
```

```python
df_raw
```

```python
#Populating month column with values from 0 to 11
months = ['January','February','March','April','May','June','July','August','September','October','November','December']
current_month = 0

for month in months:
    df_raw.loc[df_raw['relDate'].str.contains(month), 'month'] = current_month
    current_month += 1
```
